# Indexing in pandas

Pandas indexing works in one of two paradigms. The first is index-based selection: selecting data based on its numerical position in the data. iloc follows this paradigm.



```python
# selecting first row values : returns a panda series

s_1 = data.iloc[:3]
print('Type : ' ,type(s_1))
```

```python
s_1.loc[:2]
```

```python
# let's select the first 10 item's type , year , outlet type 
data.iloc[:10, [4,7,8]] # remember "10" is not included in the range
```

#### Negative numbers can be used in selection 

```python
# let's get the last 10 
data.iloc[-10:, [4,7,8]]
```

### Label based selection
In this paradigm, it's the data index value, not its position, which matters.

Both loc and iloc are row-first, column-second.

```python
# let's select the first 11 item's type , year , outlet type 
data.loc[:10, ['Item_Type' , 'Outlet_Establishment_Year' , 'Outlet_Size']] # remember in Loc "10" is included
```

### You can use specific column as an index 

```python
# we will index data by the year's column
data.set_index('Outlet_Establishment_Year')
```

# Conditional selection

select values from one column when it meets condition 

```python
# Select data in year 1987
data.Outlet_Establishment_Year == 1987
```

This operation produced a Series of True/False booleans

```python
data.loc[data.Outlet_Establishment_Year == 1987]
```

## We can use the ampersand (&) to add more conditions 

```python
# get data in 2002  with low fat

data.loc[(data.Outlet_Establishment_Year == 2002)&(data.Item_Fat_Content == 'Low Fat')]
```

```python
# exclude data from 2000  until now 
data.loc[data.Outlet_Establishment_Year.isin((data.loc[data.Outlet_Establishment_Year >= 2000]).Outlet_Establishment_Year)]
```

```python
# exclude lowfat and regular data
data.loc[data.Item_Fat_Content.isin(['Regular','Low Fat'])]
```

```python
# Get not null data in column outlet size 
data.loc[data.Outlet_Size.notnull()]
```

```python
# Highlight null or not nan data in outlet size
data.loc[data.Outlet_Size.isna()]
```

# Grouping , Aggregating and Sorting
